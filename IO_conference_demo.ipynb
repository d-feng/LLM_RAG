## Cell 4: LLM Query Interface Setup

````python
# Configure LLM API (choose one)
LLM_PROVIDER = None

# Option 1: Anthropic Claude
try:
    from anthropic import Anthropic
    ANTHROPIC_API_KEY = os.environ.get("ANTHROPIC_API_KEY")
    if ANTHROPIC_API_KEY:
        anthropic_client = Anthropic(api_key=ANTHROPIC_API_KEY)
        LLM_PROVIDER = "anthropic"
        print("‚úÖ Claude API configured (Recommended)")
    else:
        print("‚ö†Ô∏è  ANTHROPIC_API_KEY not found in environment")
except ImportError:
    print("‚ö†Ô∏è  anthropic package not installed. Run: pip install anthropic")

# Option 2: OpenAI GPT-4
if not LLM_PROVIDER:
    try:
        from openai import OpenAI
        OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
        if OPENAI_API_KEY:
            openai_client = OpenAI(api_key=OPENAI_API_KEY)
            LLM_PROVIDER = "openai"
            print("‚úÖ OpenAI API configured")
        else:
            print("‚ö†Ô∏è  OPENAI_API_KEY not found in environment")
    except ImportError:
        print("‚ö†Ô∏è  openai package not installed. Run: pip install openai")

if not LLM_PROVIDER:
    print("\n‚ùå No LLM provider configured. Please set API key:")
    print("   export ANTHROPIC_API_KEY='your-key'  # Recommended")
    print("   OR")
    print("   export OPENAI_API_KEY='your-key'")


def query_llm(prompt: str, provider: str = LLM_PROVIDER, max_tokens: int = 4000) -> str:
    """Send query to LLM with conference context"""
    
    if not provider:
        return "‚ùå LLM provider not configured. Please set API key and restart kernel."
    
    # Append conference context to prompt
    full_prompt = f"""{prompt}

=== CONFERENCE INTELLIGENCE CONTEXT ===
{json.dumps(conference_data, indent=2)}
=== END CONTEXT ===
"""
    
    try:
        if provider == "anthropic":
            response = anthropic_client.messages.create(
                model="claude-sonnet-4-5-20250929",
                max_tokens=max_tokens,
                messages=[{
                    "role": "user",
                    "content": full_prompt
                }]
            )
            return response.content[0].text
        
        elif provider == "openai":
            response = openai_client.chat.completions.create(
                model="gpt-4-turbo-preview",
                messages=[{
                    "role": "user",
                    "content": full_prompt
                }],
                max_tokens=max_tokens
            )
            return response.choices[0].message.content
    
    except Exception as e:
        return f"‚ùå Error querying LLM: {str(e)}"

print("\n‚úÖ LLM query interface ready")
print(f"üì° Using provider: {LLM_PROVIDER or 'None'}")
````

## Cell 5: Demo Query #1 - Target Matching

**Scenario:** We have a GM-CSF neutralizing antibody (Lenzilumab). Which indication should we pursue?

````python
display(Markdown("## üéØ Demo 1: Target Matching for Drug Candidate"))

query_1 = """Using the conference intelligence context provided, analyze the following drug candidate and recommend the best target match:

DRUG INFORMATION:
- Drug Name: Lenzilumab
- Mechanism of Action: Neutralizing monoclonal antibody against GM-CSF
- Current Development Stage: Phase 2/3 (completed COVID trials)
- Key Pharmacology: Blocks GM-CSF signaling, reduces inflammatory myeloid cell activation

ANALYSIS REQUIRED:
1. Which target (rank 1-7) best matches this drug's MOA?
2. What is the success score and rationale for this target?
3. What specific patient population should be targeted?
4. What biomarkers are required for patient selection?
5. What are the key risks and mitigation strategies?

Provide ranking of top 3 target matches with justification.
"""

print("üì§ Sending query to LLM...")
print("‚è±Ô∏è  This may take 20-30 seconds...\n")

response_1 = query_llm(query_1)

print("‚úÖ Response received!\n")
display(Markdown(response_1))
````

## Cell 6: Demo Query #2 - Trial Design
